```python
# ============================================================
# HOME INSURANCE
# From raw CSVs -  joined dataset -  LR / RF / CatBoost models
# + metrics, visuals, SHAP, outreach lists
# ============================================================

!pip install catboost shap -q

import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

from catboost import CatBoostClassifier, Pool

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score,
    roc_curve,
    precision_recall_curve
)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

import shap

plt.rcParams["figure.figsize"] = (7, 5)
plt.rcParams["axes.grid"] = True

OUT_DIR = "assignment_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# ============================================================
# 1. LOAD RAW DATA
# ============================================================

campaign = pd.read_csv("/content/Dummy dataset campaign.csv")
mortgage = pd.read_csv("/content/dummy dataset mortgage.csv")

print("Campaign shape:", campaign.shape)
print("Mortgage shape:", mortgage.shape)
print("Mortgage columns:", mortgage.columns.tolist())

# ============================================================
# 2. CAMPAIGN PREP (LABEL + BASIC SIGNALS)
# ============================================================

# Normalise campaign column names
campaign.columns = (
    campaign.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print("\nCampaign columns:", campaign.columns.tolist())

# 2.1 Define label: bought (1) vs not (0)
campaign["bought"] = (campaign["created_account"] == "Yes").astype(int)

print("\nLabel distribution (proportion):")
print(campaign["bought"].value_counts(normalize=True).rename("proportion"))

# interest vs purchase signal
ct_interest = pd.crosstab(
    campaign["interested_insurance"],
    campaign["bought"],
    margins=True,
    normalize="index"
)
print("\nInterest vs purchase (row-normalised):")
display(ct_interest)

# ============================================================
# 3. MORTGAGE PREP (NAMES, DOB, SALARY, EMPLOYMENT)
# ============================================================

print("\nOriginal mortgage columns:", mortgage.columns.tolist())

# 3.1 Standardise PAYE name if needed
if "paye" in mortgage.columns:
    pass
elif "Employer PAYE reference" in mortgage.columns:
    mortgage.rename(columns={"Employer PAYE reference": "paye"}, inplace=True)
elif "Employer PAYE Reference" in mortgage.columns:
    mortgage.rename(columns={"Employer PAYE Reference": "paye"}, inplace=True)
elif "employer_paye_reference" in mortgage.columns:
    mortgage.rename(columns={"employer_paye_reference": "paye"}, inplace=True)
else:
    print("⚠️ WARNING: Couldn't find a PAYE column – please check header names.")

# 3.2 Split full_name into title / first / middle / last
title_list = ["Mr.", "Mrs.", "Miss", "Ms.", "Dr.", "Mr", "Mrs", "Miss", "Ms", "Dr"]

def split_full_name(full_name):
    """
    Split 'full_name' into: name_title, first_name, middle_name, last_name.
    """
    if pd.isna(full_name) or str(full_name).strip() == "":
        return pd.Series([np.nan, np.nan, np.nan, np.nan])

    parts = str(full_name).strip().split()
    name_title = None

    if parts and parts[0] in title_list:
        name_title = parts[0]
        parts = parts[1:]

    if len(parts) == 0:
        first_name = middle_name = last_name = None
    elif len(parts) == 1:
        first_name, middle_name, last_name = parts[0], None, None
    elif len(parts) == 2:
        first_name, last_name = parts
        middle_name = None
    else:
        first_name = parts[0]
        last_name = parts[-1]
        middle_name = " ".join(parts[1:-1])

    return pd.Series([name_title, first_name, middle_name, last_name])

mortgage[["name_title", "first_name", "middle_name", "last_name"]] = (
    mortgage["full_name"].apply(split_full_name)
)

# 3.3 DOB -  approximate age
mortgage["dob_parsed"] = pd.to_datetime(
    mortgage["dob"], dayfirst=True, errors="coerce"
)
ref_date = pd.Timestamp("2024-01-01")
mortgage["age_approx"] = (
    (ref_date - mortgage["dob_parsed"]).dt.days / 365.25
)

mortgage["age_flag_implausible"] = (
    (mortgage["age_approx"] < 18) | (mortgage["age_approx"] > 100)
)

print("\nAge approx summary:")
display(mortgage["age_approx"].describe())

# 3.4 Employment duration (months)
mortgage["employment_months"] = (
    mortgage["years_with_employer"] * 12
    + mortgage["months_with_employer"]
)

print("\nEmployment duration (months) summary:")
display(mortgage["employment_months"].describe())

# 3.5 Parse salary_band -  salary_type, annual_salary_gbp
def parse_salary_band(s):
    """
    Parse salary_band into:
      - salary_type: yearly / monthly / weekly / range_yearly_midpoint /
                     foreign_currency_or_other / unknown_format / missing
      - annual_salary_gbp: float or NaN
    """
    if pd.isna(s):
        return pd.Series(["missing", np.nan])

    s = str(s).strip()
    if not s.startswith("£"):
        # e.g. foreign currency, or text -  we treat as unknown FX
        return pd.Series(["foreign_currency_or_other", np.nan])

    s_clean = s[1:].strip()  # drop leading £

    # Yearly salary
    m_yearly = re.match(r"^([\d,\.]+)\s*yearly$", s_clean, re.IGNORECASE)
    if m_yearly:
        val = float(m_yearly.group(1).replace(",", ""))
        return pd.Series(["yearly", val])

    # Monthly salary
    m_month = re.match(r"^([\d,\.]+)\s*per\s*month$", s_clean, re.IGNORECASE)
    if m_month:
        val = float(m_month.group(1).replace(",", ""))
        return pd.Series(["monthly", val * 12.0])

    # Weekly salary
    m_week = re.match(r"^([\d,\.]+)\s*pw$", s_clean, re.IGNORECASE)
    if m_week:
        val = float(m_week.group(1).replace(",", ""))
        return pd.Series(["weekly", val * 52.0])

    # Yearly range
    m_range = re.match(
        r"^([\d,\.]+)\s*-\s*([\d,\.]+)\s*range$", s_clean, re.IGNORECASE
    )
    if m_range:
        low = float(m_range.group(1).replace(",", ""))
        high = float(m_range.group(2).replace(",", ""))
        midpoint = (low + high) / 2.0
        return pd.Series(["range_yearly_midpoint", midpoint])

    return pd.Series(["unknown_format", np.nan])

mortgage[["salary_type", "annual_salary_gbp"]] = mortgage["salary_band"].apply(
    parse_salary_band
)

print("\nSalary type value counts:")
print(mortgage["salary_type"].value_counts())

print("\nAnnual salary (GBP, approx) summary:")
display(mortgage["annual_salary_gbp"].describe())

# 3.6 Optional numeric summary for documentation
numeric_cols_mort = mortgage.select_dtypes(include=[np.number]).columns.tolist()
mortgage_numeric_summary = mortgage[numeric_cols_mort].describe().T
mortgage_numeric_summary.to_csv(os.path.join(OUT_DIR, "mortgage_numeric_summary.csv"))

print("\nNumeric columns in mortgage:")
print(numeric_cols_mort)

# ============================================================
# 4. JOIN CAMPAIGN + MORTGAGE (NAME-BASED)
# ============================================================

def normalise_name(s):
    if pd.isna(s):
        return None
    return str(s).strip().lower()

campaign["first_name_norm"] = campaign["first_name"].apply(normalise_name)
campaign["last_name_norm"]  = campaign["last_name"].apply(normalise_name)

mortgage["first_name_norm"] = mortgage["first_name"].apply(normalise_name)
mortgage["last_name_norm"]  = mortgage["last_name"].apply(normalise_name)

# Restrict mortgage to unique (first,last) to avoid 1 to many joins
name_counts = (
    mortgage.groupby(["first_name_norm", "last_name_norm"])
    .size()
    .reset_index(name="n_mort_rows")
)
unique_names = name_counts[name_counts["n_mort_rows"] == 1][["first_name_norm", "last_name_norm"]]

mortgage_unique = mortgage.merge(
    unique_names,
    on=["first_name_norm", "last_name_norm"],
    how="inner"
)

print("\nMortgage rows (original):", len(mortgage))
print("Mortgage rows with unique (first,last):", len(mortgage_unique))

# Join campaign to mortgage_unique
model_df = campaign.merge(
    mortgage_unique,
    on=["first_name_norm", "last_name_norm"],
    how="left",
    suffixes=("_camp", "_mort")
)

print("\nJoined model_df shape:", model_df.shape)

# PAYE coverage check
if "paye" in model_df.columns:
    print(f"PAYE coverage after join: {model_df['paye'].notna().mean():.2%}")
else:
    print("⚠️ PAYE column not present in model_df – check earlier steps.")

# Rename campaign name columns back to simple form
rename_map = {}
for col in ["first_name", "last_name", "name_title"]:
    camp_col = f"{col}_camp"
    if camp_col in model_df.columns:
        rename_map[camp_col] = col

model_df.rename(columns=rename_map, inplace=True)

# Quick sanity sample
cols_check = [
    "first_name", "last_name",
    "first_name_mort", "last_name_mort",
    "age", "postcode",
    "town", "paye", "salary_band", "annual_salary_gbp",
    "years_with_employer", "employment_months",
    "new_mortgage", "sex", "religion", "race",
    "demographic_characteristic", "bought"
]
cols_check = [c for c in cols_check if c in model_df.columns]

print("\nSample of enriched campaign rows:")
display(model_df[cols_check].head(10))

# Save joined dataset for traceability
model_df.to_csv(os.path.join(OUT_DIR, "campaign_mortgage_model_dataset.csv"), index=False)
print("\n✅ Final modelling dataset saved as: campaign_mortgage_model_dataset.csv")

# For downstream modelling, use df = model_df
df = model_df.copy()

# ============================================================
# 5. TARGET & FEATURES
# ============================================================

# Target column
if "bought" not in df.columns:
    df["bought"] = (df["created_account"] == "Yes").astype(int)
y = df["bought"]

# Drop columns not suitable as features
drop_cols = [
    "bought", "created_account",
    "sex", "race", "religion", "relationship", "native_country",
    "participant_id", "company_email",
    "full_name", "first_name", "last_name", "name_title",
    "first_name_mort", "last_name_mort", "name_title_mort", "middle_name",
    "dob", "dob_parsed", "age_approx", "age_flag_implausible",
    "salary_band", "first_name_norm", "last_name_norm"
]
drop_cols = [c for c in drop_cols if c in df.columns]

X = df.drop(columns=drop_cols)

print("\nRemaining feature columns:", X.columns.tolist())
print("Target positive rate:", y.mean())

# Identify categorical vs numeric
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()
num_cols = X.select_dtypes(exclude=["object"]).columns.tolist()

print("\nCategorical features:", cat_cols)
print("\nNumeric features:", num_cols)

# ============================================================
# 6. BASIC EDA - CLASS BALANCE & MISSINGNESS
# ============================================================

# 6.1 Class distribution
class_counts = y.value_counts().sort_index()
labels = ["Not bought (0)", "Bought (1)"]
values = class_counts.values

plt.figure()
plt.bar(labels, values)
for i, v in enumerate(values):
    plt.text(i, v + 50, f"{v} ({v/len(y):.2%})", ha="center", va="bottom")
plt.title("Class Distribution: Home Insurance Purchase")
plt.ylabel("Number of customers")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "class_distribution.png"))
plt.show()

# 6.2 Missingness by column (top 20)
missing_pct = X.isna().mean().sort_values(ascending=False) * 100
print("\nMissingness (% of rows) – top 20:")
display(missing_pct.head(20))

plt.figure(figsize=(8, 5))
missing_pct.head(20).plot(kind="barh")
plt.xlabel("% missing")
plt.title("Top 20 Features by Missingness")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "missingness_top20.png"))
plt.show()

# ============================================================
# 7. TRAIN / TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nTrain shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train positive rate:", y_train.mean(), "Test positive rate:", y_test.mean())

# ============================================================
# 8. METRIC & PLOT HELPERS
# ============================================================

def precision_at_k(y_true, proba, k=0.01):
    """Precision in the top k fraction of customers."""
    n = len(proba)
    cutoff = max(1, int(n * k))
    order = np.argsort(-proba)
    top_k = y_true.iloc[order[:cutoff]]
    return top_k.mean()


def evaluate_ranking_model(name, y_true, proba, out_prefix):
    """Compute AUROC, AUPRC, Precision@K, Lift and produce ROC/PR/Lift plots."""
    base_rate = y_true.mean()
    auroc = roc_auc_score(y_true, proba)
    auprc = average_precision_score(y_true, proba)

    print(f"\n===== {name} =====")
    print(f"AUROC: {auroc:.4f}")
    print(f"AUPRC: {auprc:.4f}")
    print(f"Base positive rate: {base_rate:.4f}")

    for k in [0.01, 0.05, 0.10]:
        p_at_k = precision_at_k(y_true, proba, k)
        lift_k = p_at_k / base_rate
        print(f"Top {int(k*100)}% -  Precision={p_at_k:.4f}, Lift={lift_k:.2f}")

    # ROC curve
    fpr, tpr, _ = roc_curve(y_true, proba)
    plt.figure()
    plt.plot(fpr, tpr, label=f"{name} (AUROC={auroc:.3f})")
    plt.plot([0, 1], [0, 1], linestyle="--", label="Random")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve – {name}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, f"{out_prefix}_roc.png"))
    plt.show()

    # Precision–Recall curve
    prec, rec, _ = precision_recall_curve(y_true, proba)
    plt.figure()
    plt.plot(rec, prec, label=f"{name} (AUPRC={auprc:.3f})")
    plt.hlines(base_rate, 0, 1, linestyles="--", label=f"Baseline ({base_rate:.3f})")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"Precision–Recall Curve – {name}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, f"{out_prefix}_pr.png"))
    plt.show()

    # Lift curve
    ks = [0.01, 0.05, 0.10, 0.20, 0.30]
    lifts = []
    for k in ks:
        p_k = precision_at_k(y_true, proba, k)
        lifts.append(p_k / base_rate if base_rate > 0 else np.nan)

    plt.figure()
    plt.plot([k*100 for k in ks], lifts, marker="o")
    plt.axhline(1.0, linestyle="--", label="Random = 1.0")
    plt.xlabel("Top K% of customers")
    plt.ylabel("Lift vs random")
    plt.title(f"Lift Curve – {name} (Test Set)")
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, f"{out_prefix}_lift.png"))
    plt.show()

    return auroc, auprc

# ============================================================
# 9. LOGISTIC REGRESSION (GOVERNANCE BASELINE)
# ============================================================

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess_lr = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols)
    ]
)

log_reg = Pipeline(steps=[
    ("preprocess", preprocess_lr),
    ("model", LogisticRegression(
        max_iter=200,
        class_weight="balanced",
        n_jobs=-1
    ))
])

log_reg.fit(X_train, y_train)
proba_lr = log_reg.predict_proba(X_test)[:, 1]
lr_auc, lr_ap = evaluate_ranking_model("Logistic Regression", y_test, proba_lr, "logreg")

# ============================================================
# 10. RANDOM FOREST (STRONG, NON-LINEAR BAGGING BASELINE)
# ============================================================

preprocess_rf = preprocess_lr  # reuse same preprocessor

rf_model = Pipeline(steps=[
    ("preprocess", preprocess_rf),
    ("model", RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=5,
        n_jobs=-1,
        class_weight="balanced_subsample",
        random_state=42
    ))
])

rf_model.fit(X_train, y_train)
proba_rf = rf_model.predict_proba(X_test)[:, 1]
rf_auc, rf_ap = evaluate_ranking_model("Random Forest", y_test, proba_rf, "random_forest")

# ============================================================
# 11. CATBOOST (CHAMPION MODEL – TREE BOOSTING ON CATEGORICALS)
# ============================================================

# CatBoost works directly on raw categoricals – different pre-processing path
X_cb = X.copy()
for col in cat_cols:
    X_cb[col] = X_cb[col].astype("object").fillna("MISSING")

X_train_cb, X_test_cb, y_train_cb, y_test_cb = train_test_split(
    X_cb, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

cat_feature_indices = [X_cb.columns.get_loc(c) for c in cat_cols]

train_pool = Pool(X_train_cb, label=y_train_cb, cat_features=cat_feature_indices)
test_pool = Pool(X_test_cb, label=y_test_cb, cat_features=cat_feature_indices)

model_cb = CatBoostClassifier(
    loss_function="Logloss",
    eval_metric="AUC",
    depth=6,
    learning_rate=0.05,
    iterations=1000,
    random_seed=42,
    auto_class_weights="Balanced",
    verbose=100
)

model_cb.fit(train_pool, eval_set=test_pool, use_best_model=True)

proba_cb = model_cb.predict_proba(test_pool)[:, 1]
cb_auc, cb_ap = evaluate_ranking_model("CatBoost", y_test_cb, proba_cb, "catboost")

# ============================================================
# 12. MODEL COMPARISON TABLE
# ============================================================

comparison = pd.DataFrame({
    "model": ["Logistic Regression", "Random Forest", "CatBoost"],
    "AUROC": [lr_auc, rf_auc, cb_auc],
    "AUPRC": [lr_ap, rf_ap, cb_ap]
})
print("\nModel comparison:")
display(comparison)
comparison.to_csv(os.path.join(OUT_DIR, "model_comparison.csv"), index=False)

# ============================================================
# 13. CATBOOST FEATURE IMPORTANCE (GLOBAL)
# ============================================================

feature_importances = model_cb.get_feature_importance(train_pool)
fi_df = pd.DataFrame({
    "feature": X_cb.columns,
    "importance": feature_importances
}).sort_values("importance", ascending=False)

print("\nTop 20 CatBoost feature importances:")
display(fi_df.head(20))
fi_df.to_csv(os.path.join(OUT_DIR, "catboost_feature_importance.csv"), index=False)

plt.figure(figsize=(8, 6))
top_n = 15
plt.barh(fi_df["feature"].head(top_n)[::-1],
         fi_df["importance"].head(top_n)[::-1])
plt.xlabel("Importance")
plt.title("CatBoost Feature Importance (Top 15)")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "catboost_feature_importance_top15.png"))
plt.show()

# ============================================================
# 14. CATBOOST SHAP VALUES (GLOBAL SUMMARY)
# ============================================================

shap_values_cb = model_cb.get_feature_importance(
    train_pool,
    type="ShapValues"
)
shap_values = shap_values_cb[:, :-1]  # last column = expected value
print("SHAP matrix shape:", shap_values.shape)

sample_size = min(2000, X_train_cb.shape[0])
idx_sample = np.random.choice(X_train_cb.shape[0], size=sample_size, replace=False)
X_shap = X_train_cb.iloc[idx_sample]
shap_vals_sample = shap_values[idx_sample, :]

shap.summary_plot(
    shap_vals_sample,
    X_shap,
    feature_names=X_cb.columns,
    show=False
)
plt.title("SHAP Summary – CatBoost (sample)")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "catboost_shap_summary.png"))
plt.show()

# ============================================================
# 15. PER-CUSTOMER SHAP REASON CODES & OUTREACH LISTS
# ============================================================

def build_reason_codes_from_shap(shap_values_all, feature_names, top_n=5):
    reasons = []
    for i in range(shap_values_all.shape[0]):
        row = shap_values_all[i]
        idx = np.argsort(-np.abs(row))[:top_n]
        codes = []
        for j in idx:
            direction = "↑" if row[j] > 0 else "↓"
            codes.append(f"{feature_names[j]} {direction} ({row[j]:+.2f})")
        reasons.append("; ".join(codes))
    return reasons

# Score full dataset with CatBoost
X_cb_full = X_cb.copy()
full_pool = Pool(X_cb_full, cat_features=cat_feature_indices)
proba_full_cb = model_cb.predict_proba(full_pool)[:, 1]

# SHAP for all rows (sample for production)
shap_values_all = model_cb.get_feature_importance(
    full_pool,
    type="ShapValues"
)[:, :-1]

reason_codes = build_reason_codes_from_shap(shap_values_all, X_cb_full.columns, top_n=5)

scores_cb = pd.DataFrame({
    "cb_proba": proba_full_cb,
    "bought": y.values,
    "top5_reason_codes": reason_codes
}, index=df.index)

scores_cb = scores_cb.sort_values("cb_proba", ascending=False)
scores_cb["rank"] = np.arange(1, len(scores_cb) + 1)
scores_cb["percentile"] = scores_cb["rank"] / len(scores_cb)
scores_cb["top_1pct"] = scores_cb["percentile"] <= 0.01
scores_cb["top_5pct"] = scores_cb["percentile"] <= 0.05
scores_cb["top_10pct"] = scores_cb["percentile"] <= 0.10

# Save key outputs
scores_cb.to_csv(os.path.join(OUT_DIR, "catboost_scored_with_reason_codes.csv"), index=False)
scores_cb[scores_cb["top_1pct"]].to_csv(os.path.join(OUT_DIR, "catboost_top1pct_outreach.csv"), index=False)
scores_cb[scores_cb["top_5pct"]].to_csv(os.path.join(OUT_DIR, "catboost_top5pct_outreach.csv"), index=False)
scores_cb[scores_cb["top_10pct"]].to_csv(os.path.join(OUT_DIR, "catboost_top10pct_outreach.csv"), index=False)

top10_reasons = scores_cb[["cb_proba", "bought", "top5_reason_codes"]].head(10)
top10_reasons.to_csv(os.path.join(OUT_DIR, "top10_with_reason_codes_for_slide.csv"), index=False)

print("\n All outputs saved to:", OUT_DIR)
```
